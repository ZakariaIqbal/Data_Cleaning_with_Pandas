## Paying Customer
Standardizing Column Values Using Replace 

```python
df.head(5)
```

```python
#for Paying Customer Column
df['Paying Customer']=  df['Paying Customer'].str.replace('Yes','Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No','N')

#for  Customer Column
df['Do_Not_Contact']=  df['Do_Not_Contact'].str.replace('Yes','Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No','N')

df[['Paying Customer','Do_Not_Contact']].head()
```

# Fill Null Values

```python
df = df.fillna('')
df.head()
```

## Filtering  Down Rows of Data

We dont Want to Givethe call center numbers that or people who don't have numbers so we want to remove those.


```python
for x in df.index:
   if df.loc[x,'Do_Not_Contact'] == 'Y':
       df.drop(x,inplace =True)
df.head()
```

```python
# Removing null phone_number
for x in df.index:
   if df.loc[x,'Phone_Number'] == '':
       df.drop(x,inplace =True)


# Another way to Drop Null Values
#df.dropna(subset='Phone_Number',inplace =True)

df
```
